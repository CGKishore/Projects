## About the Data 

#### year:
The year in which the air quality measurements were taken.

#### month:
The month of the year in which the measurements were taken.

#### day:
The day of the month in which the measurements were taken.

#### hour:
The hour of the day (in 24-hour format) at which the measurements were taken.

#### PM2.5:
The concentration of particulate matter (PM) with a diameter of 2.5 micrometers or less, measured in micrograms per cubic meter (µg/m³). PM2.5 can penetrate deep into the lungs and is associated with various health problems.


#### PM10:
The concentration of particulate matter with a diameter of 10 micrometers or less, measured in micrograms per cubic meter (µg/m³). PM10 includes dust, pollen, and mold spores, which can cause respiratory issues.

#### SO2:
The concentration of sulfur dioxide in the air, measured in parts per billion (ppb) or micrograms per cubic meter (µg/m³). SO2 is a significant pollutant that can affect air quality and human health.

#### NO2:
The concentration of nitrogen dioxide in the air, typically measured in parts per billion (ppb) or micrograms per cubic meter (µg/m³). NO2 can cause respiratory problems and is a component of smog.

#### CO:
The concentration of carbon monoxide in the air, measured in parts per million (ppm) or micrograms per cubic meter (µg/m³). CO is a colorless, odorless gas that can be harmful when inhaled in large amounts.

#### O3:
The concentration of ozone in the air, typically measured in parts per billion (ppb) or micrograms per cubic meter (µg/m³). Ozone at ground level can cause respiratory issues and is a key component of smog.

#### TEMP:
The temperature at the time of measurement, usually in degrees Celsius (°C). Temperature can influence air quality and pollutant levels.

#### PRES:
The atmospheric pressure, typically measured in hPa (hectopascals) or mmHg (millimeters of mercury). Changes in pressure can affect weather conditions and air quality.

#### DEWP:
The dew point temperature, which indicates the temperature at which air becomes saturated with moisture. This is important for understanding humidity and potential for fog or precipitation.

#### RAIN:
The amount of rain recorded during the measurement period, usually measured in millimeters (mm). Rain can affect air quality by washing away pollutants.

#### wd:
The wind direction, typically measured in cardinal directions (e.g., N, S, E, W) or in degrees. Wind direction can influence the dispersion of air pollutants.

#### WSPM:
Wind speed, often measured in meters per second (m/s) or kilometers per hour (km/h). Wind speed can affect the dispersion of air pollutants and overall air quality.

#### station:
The name or identifier of the monitoring station where the measurements were taken. This can help in identifying the location of the air quality data.


## Before moving further first we have to what is Air quality index? how to calculate it?

#### What is the Air Quality Index (AQI)? 
Air Quality Index (AQI) is a number used to convey the quality of air by the government to the general public. Air quality deteriorates with an increase in the concentration of pollutants. The Air Quality Index represents the severity of pollution for ordinary people

AQI range is from 0-500, from 0 being good and 500 being severe. There are eight major pollutants to be taken into account for AQI calculation, viz. particulate matter (PM 10 and PM 2.5), carbon monoxide (CO), ozone (O3), nitrogen dioxide (NO2), sulfur dioxide (SO2), ammonia (NH3), and lead (Pb). To calculate AQI, data for a minimum of three pollutants must be present, of which one should be either PM10 or PM2.5, AQI ranging from 0-500 has different concentrations for each pollutant and has health effects accordingly..

### AQI Ranges and Descriptions
Good (0 - 50): Air quality is considered satisfactory, and air pollution poses little or no risk. In this range, the air quality is ideal for outdoor activities, and no health impacts are expected.

Moderate (51 - 100): Air quality is acceptable; however, there may be a concern for some pollutants for a small number of people. Generally, this means that minor health effects may be experienced by sensitive individuals, such as those with respiratory conditions. The general public is unlikely to be affected.

Unhealthy for Sensitive Groups (101 - 150): Members of sensitive 
groups, including children, the elderly, and people with respiratory or heart conditions, may experience health effects. However, the general public is unlikely to be affected. It’s advisable for sensitive individuals to limit prolonged outdoor exertion.

Unhealthy (151 - 200): At this level, everyone may begin to experience health effects. Sensitive groups may experience more serious health effects. It’s recommended that everyone limit outdoor activities, particularly strenuous ones.

Very Unhealthy (201 - 300): This range signifies a health alert where everyone may experience more serious health effects. Sensitive groups are at a greater risk and should avoid prolonged outdoor exertion. It’s advisable for everyone to minimize outdoor activities.

Hazardous (301 - 500): This level indicates emergency conditions, where health warnings are issued. The entire population is more likely to be affected, with serious health effects likely for all individuals. It is crucial to stay indoors, and outdoor activities should be avoided altogether.



## How to calculate AQI

##### Step 1:

Identify Key Pollutants: The AQI is typically calculated for the following pollutants:

- PM2.5 (Particulate Matter)
- PM10 (Particulate Matter)
- Ozone (O3)
- Nitrogen Dioxide (NO2)
- Sulfur Dioxide (SO2)
- Carbon Monoxide (CO)

##### Step2:

Obtain Concentration Values: Collect the concentration data for each pollutant from air quality monitoring stations. These values are usually measured in micrograms per cubic meter (µg/m³) for particulate matter and in parts per billion (ppb) or parts per million (ppm) for gases.

##### Step 3:

Convert Concentration to AQI: Use standard conversion formulas to calculate the AQI for each pollutant. Each pollutant has its own set of breakpoints that relate the concentration to an AQI value. The following is a common approach:


The AQI formula is given by:

$$
\text{AQI} = \frac{(I_{\text{high}} - I_{\text{low}})}{(C_{\text{high}} - C_{\text{low}})} \times (C - C_{\text{low}}) + I_{\text{low}}
$$
- AQI: The calculated Air Quality Index for the pollutant.
- C: The measured concentration of the pollutant.
- C_{low}: The lower concentration breakpoint that corresponds to the AQI range.
- C_{high}: The upper concentration breakpoint that corresponds to the AQI range.
- I_{low}: The lower AQI value corresponding to the breakpoints.
- I_{high}: The upper AQI value corresponding to the breakpoints.

```python
import pandas as pd


# Function to calculate AQI for each pollutant
def calculate_aqi_pm25(pm25):
    if pm25 >=0 and pm25 <= 12:
        return ((50 - 0) / (12 - 0)) * ( pm25 - 0) + 0
    elif pm25 >= 13 and pm25 <= 35:
        return ((100 - 50) / (35 - 12)) * (pm25 - 12) + 50
    elif pm25 >= 36 and pm25 <= 55:
        return ((150 - 100) / (55 - 35)) * (pm25 - 35) + 100
    elif pm25 >= 56 and pm25 <= 150:
        return ((200 - 150) / (150 - 55)) * (pm25 - 55) + 150
    elif pm25 >= 151 and pm25 <= 250:
        return ((300 - 200) / (250 - 150)) * (pm25 - 150) + 200
    elif pm25 >= 251:
        return ((500 - 300) / ( 500 - 250)) * (pm25 -250) +  300
   


def calculate_aqi_pm10(pm10):
    if pm10 >= 0 and pm10 <= 54:
        return ((50 - 0) / (54 - 0)) * (pm10 - 0) + 0
    elif pm10 >= 55 and pm10 <= 154:
        return ((100 - 50) / (154 - 54)) * (pm10 - 55) + 50
    elif pm10 >= 155 and pm10 <= 254:
        return ((150 - 100) / (254 - 154)) * (pm10 - 155) + 100
    elif pm10 >= 255 and pm10 <= 354:
        return ((200 - 150) / (354 - 254)) * (pm10 - 255) + 150
    elif pm10 >= 355 and pm10 <= 424:
        return ((300 - 200) / (425 - 355)) * (pm10 - 355) + 200
    elif pm10 >= 425:
        return ((500 - 300) / (604 -425)) * (pm10 - 425) + 300
    
    
    

def calculate_aqi_so2(so2):
    if so2>=0 and so2 <= 150:
        return ((50 - 0)/(150 -0)) * (so2 - 0) + 0
    elif so2>= 151 and so2 <= 500:
        return ((100 - 50) / (500 - 151)) * (so2 - 151) + 50
    elif s02 >= 501 and so2 <= 650:
        return ((150 - 100) / (600 - 501)) * (so2 - 501) + 100
    elif s02 >=651 and so2 <= 800:
        return ((200 - 150) / (800 - 651)) * (so2 - 651) + 150
    

def calculate_aqi_no2(no2):
    if no2<=0  and no2 <= 100:
        return ((50 - 0) / (100 -0)) * (no2-0) + 0
    elif no2<=101 and no2<= 200:
        return ((100-50) / (200 - 100)) * (no2 - 100) + 50
    elif no2 <=201 and no2<=700:
        return ((150-100) / (700 - 1200)) * (no2 - 200) + 100
    elif no2>= 701 and no2 <= 1200:
        return ((200 - 150) / (1200 - 700)) * (no2 - 700) + 200
    elif no2>=1201 and no2 <= 2340:
        return ((300 - 200) / (2340 - 1200)) * (no2- 1200) + 300
    elif no2>=2341 and no2<=3090:
        return ((400-300) / (3090 - 2340)) * ( no2 - 2340)+ 400
    elif no2 <= 3091 and no2>=3840:
        return ((500 -400)/ (3840 - 3090)) * (no2 -3090)+500
        

def calculate_aqi_co(co):
    if co>= 0 and co <= 5:
        return ((50-0)/(5-0)) * (co -0) +0
    elif co>=5.1 and co <= 10:
        return ((100-50) / (10-5)) * (co - 5) + 50
    elif co>= 10.1 and co <= 35:
        return ((150-100) / (35-10)) * (co - 10) + 100
    elif co>= 36.1 and co <= 60:
        return ((200-150) / (60-35)) * (co - 35) + 200
    elif co>= 60.1 and co <= 90:
        return ((300-200) / (90-60)) * (co - 60) + 300
    elif co>= 90.1 and co<= 120:
        return ((400-300) /(120-90)) *(co - 120) + 400
    elif co>= 120.1 and co<=150:
        return ((500-400) /(150-120)) *(co - 150) + 500

def calculate_aqi_o3(o3):
    if o3 >=0 and o3 <= 160:
        return ((50-0) / (160-0)) * (o3 - 0) +0
    elif o3 >=161 and o3 <= 200:
        return ((100-50) / (200-160)) * (o3 - 160) + 50
    elif o3>=201 and o3 <= 300:
        return ((150-100) / (300-200)) * ( o3- 200) + 100
    elif o3>= 301 and o3 <= 400:
        return ((200-150) / (400-300)) * (o3- 300) + 200
    elif o3 >= 401 and o3 <= 800:
        return ((300-200) / (800-400)) * (o3 - 400) + 300
    elif o3>=801 and o3 <= 1000:
        return ((400-300)/(1000-800))*(o3 -800) + 400
    elif o3>=1001 and o3 <= 1200:
        return ((500-400) /(1200-1000))*(o3 -1000) + 500

# Calculate AQI for each pollutant

df['AQI_PM2.5'] = df['PM2.5'].apply(calculate_aqi_pm25)
df['AQI_PM10'] = df['PM10'].apply(calculate_aqi_pm10)
df['AQI_SO2'] = df['SO2'].apply(calculate_aqi_so2)
df['AQI_NO2'] = df['NO2'].apply(calculate_aqi_no2)
df['AQI_CO'] = df['CO'].apply(calculate_aqi_co)
df['AQI_O3'] = df['O3'].apply(calculate_aqi_o3)


df['Overall_AQI'] = df[['AQI_PM2.5', 'AQI_PM10', 'AQI_SO2', 'AQI_NO2', 'AQI_CO', 'AQI_O3']].mean(axis=1)


def categorize_aqi(aqi):
    if aqi >= 0 and aqi <= 50:
        return 'Good'
    elif aqi > 50 and aqi <= 100:
        return 'Moderate'
    elif aqi >100 and aqi <= 150:
        return 'Unhealthy for Sensitive Groups'
    elif aqi >150 and aqi <= 200:
        return 'Unhealthy'
    elif aqi > 200 and aqi <= 300:
        return 'Very Unhealthy'
    else:
        return 'Hazadous'


df['AQI_Category'] = df['Overall_AQI'].apply(categorize_aqi)


print(df[['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3', 'Overall_AQI', 'AQI_Category']])
```

```python
def build_model(hp):
    act=hp.Choice("activation",values=["relu","tanh"])
    model=Sequential()
    model.add(Input(shape=(6,)))
    for i in range(hp.Int("num_layers",min_value=1,max_value=5)):
        model.add(Dense(units=hp.Int(f'units_{i}',min_value=1,max_value=128),activation=act))
    model.add(Dense(1,activation="linear"))
    model.compile(optimizer="adam",loss=["mean_absolute_error"])
    return model
```

```python
tuner= keras_tuner.RandomSearch(build_model,objective="val_loss",directory="air3",project_name="see")
tuner.search(x_train,y_train,validation_data=[x_test,y_test],epochs=10,batch_size=500)
```
